### Import libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#Incorporate these libraries to run on jupyter/colab

import plotly.offline as pyo
pyo.init_notebook_mode(connected = True)

pd.options.plotting.backend = 'plotly'

#Make it Colab Friendly
import plotly.io as pio
pio.renderers.default = "colab"
```

### Choose stocks and assign a variable inside the [ ] list (tickers should match that of Yahoo Finance)

```python
tickers = ['RBLX', 'META', 'U', 'SE', 'ATVI']
```

### Set period to extract data from

```python
start = '2019-01-01'
end = dt.datetime.now()
```

### Try extracting stock data from these time period into a dataframe

```python
data = yf.download(tickers, start = start, group_by='ticker')
data.head()
```

### Dropna or FillNa to clean data

```python
data_1 = data.fillna(method='bfill')
```

```python
data_1.head()
```

### Understand and investigate the dataframe

```python
data_1.columns
```

```python
data_1.index
```

### Extract data by ticker

#### What if I wanted to see all stocks' historical closing price?

### To extract specific columns under each stock, say all "Close" columns for selected stocks, from a multi-index dataframe use the following code: 

```python
close = data_1.iloc[:, data_1.columns.get_level_values(1)=='Close']
```

```python
close
```

### Remove the second header row "Close"

```python
close = close.droplevel(1, axis =1)
close
```

```python
close.describe()
```

```python
###note that count above is 857-856 because of backfill method, if you want to see actual total trading days for the stock, use data as the dataframe instead of data_1 above when extracting columns
close_actual = data.iloc[:, data.columns.get_level_values(1)=='Close']
```

```python
close_actual = close_actual.droplevel(1, axis =1)
close_actual
```

### Using plotly to visualize data

```python
close_actual.plot()
#Compare it with close.plot() to see whether backfill is a good method. If dropna is used instead of fillna, large amount of data will be removed and the starting period of below chart will be around July 2021.
```

#### Customizing chart layout

```python
fig = close_actual.plot(title = 'Metaverse Stocks Price Movement')
fig.update_layout(legend_title_text = 'Stocks')
fig.update_xaxes(title_text='Time')
fig.update_yaxes(title_text='Price (US$')
fig.show()
```

#### Visualizing individual stock

```python
##Define a function to generate chart for stock chosen
```

```python
def showplots():
  while True: 
    try:
      uchoose = input("Enter stock ticker: ")
      fig = close_actual[uchoose].plot(title = "Stock Price Movements (YTD)")
      fig.update_layout(legend_title_text = 'Stock')
      fig.show()
    except Exception: 
      print("Sorry, stock you entered is not avaliable.")
    break
```

```python
showplots()
```
